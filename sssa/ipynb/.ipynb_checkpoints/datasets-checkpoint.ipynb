```python
def filter_abstraction(nouns, abstract_nouns):
    abstract_nouns = set(abstract_nouns)
    is_abstract = []
    for c in nouns:
        is_abstract.append(c in abstract_nouns)
    nouns = np.array(nouns)[~np.array(is_abstract)].tolist()
    return nouns
```

### build vocabulary

```python
hypo = lambda s: s.hyponyms()
hyper = lambda s: s.hypernyms()
get_name = lambda x: x.name().split('.')[0] ### get synset name from @synset
get_name_total = lambda x: x.name() ### get synset total name from @synset
get_offset = lambda ss: ss.pos()+str(ss.offset()).zfill(8)
get_def = lambda x: x.definition() ### get synset name from @synset

nouns = [x for x in wn.all_synsets(pos=wn.NOUN)] ### all noun in wordnet

# nouns = filter_abstraction(nouns, all_abstractions)

### vocab synsets
vocab_synsets = [x for x in nouns]

### vocab synset ids
get_offset = lambda ss: ss.pos()+str(ss.offset()).zfill(8)
vocab_ids = [get_offset(x) for x in vocab_synsets]

### vocab synset names
vocab_names = [get_name(x) for x in vocab_synsets]

### vocab parents
mapping_ids_idx = dict(zip(vocab_ids, range(len(vocab_ids))))
vocab_parents = []
for x in vocab_synsets:
    curr_parents = set([])
    for y in x.closure(hyper, depth=10000):
        synset_id = get_offset(y)
        if synset_id in mapping_ids_idx:
            curr_parents.add(mapping_ids_idx[synset_id])
    vocab_parents.append(curr_parents)
    
### vocab def
vocab_def = [get_def(x) for x in vocab_synsets]
```

```python
len(vocab_ids), len(vocab_names), len(vocab_parents), len(vocab_def)
```

```python
vocab = {
    'synsets': list(map(lambda x: x.name(), vocab_synsets)),
    'ids': vocab_ids,
    'names': vocab_names,
    'parents': vocab_parents,
    'def': vocab_def,
}
```

### build graph

```python
import networkx as nx
from networkx.algorithms.traversal.breadth_first_search import bfs_predecessors, bfs_successors
from networkx.algorithms.traversal.depth_first_search import dfs_predecessors, dfs_successors
from copy import deepcopy
from networkx.drawing.nx_pydot import write_dot

def create_graph(nouns, vocab_ids, vocab_names, vocab_def):
    G = nx.DiGraph()
    for i, n in enumerate(nouns):
        G.add_node(n.name(),
                   synid=n.name(),
                   id=vocab_ids[i],
                   classname=vocab_names[i],
                   definition=vocab_def[i],
                   depth=n.max_depth(),
                  )

    queue = [nouns[0]]
    while len(queue):
        n = queue.pop(0)
        children = n.hyponyms()
        for ch in children:
            G.add_edge(n.name(), ch.name())
        queue.extend(children)
    return G


def predecessor_set(G, source):
    queue = list(G.predecessors(source))
    result_set = set(queue)
    while len(queue):
        item = queue.pop(0)
        if item in G.nodes:
            pre = list(G.predecessors(item))
            queue.extend(pre)
            result_set |= set(pre)
    return result_set
```

```python
G = create_graph(nouns, vocab_ids, vocab_names, vocab_def)
```

```python
from transformers import BertTokenizer, BertModel
import torch

def initialize_bert(device='cpu'):
    # Load pre-trained BERT tokenizer and model
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')
    model.to(device)
    return model, tokenizer


@torch.no_grad()
def extract_embedding(text, model, tokenizer, device='cpu', normalize=True):
    # Tokenize the input text and add the [CLS] and [SEP] tokens
    tokens = ['[CLS]'] + tokenizer.tokenize(text) + ['[SEP]']
    # Convert the tokens to input IDs and create a tensor
    input_ids = torch.tensor([tokenizer.convert_tokens_to_ids(tokens)]).to(device)
    with torch.no_grad():
        output = model(input_ids)
    embedding = output.last_hidden_state[0][0]
    if normalize:
        embedding = F.normalize(embedding, dim=-1)
    return embedding

device = 'cuda:3'
model, tokenizer = initialize_bert(device)
```

```python
embedding1 = extract_embedding('winter_squash: any of various plants of the species Cucurbita maxima and Cucurbita moschata producing squashes that have hard rinds and mature in the fall', model, tokenizer, device=device)
embedding2 = extract_embedding('winter_squash: any of various fruits of the gourd family with thick rinds and edible yellow to orange flesh that mature in the fall and can be stored for several months', model, tokenizer, device=device)
embedding3 = extract_embedding('summer_squash: any of various usually bushy plants producing fruit that is eaten while immature and before the rind or seeds harden', model, tokenizer, device=device)
embedding4 = extract_embedding('summer_squash: any of various fruits of the gourd family that mature during the summer; eaten while immature and before seeds and rind harden', model, tokenizer, device=device)

embedding = torch.stack([embedding1, embedding2, embedding3, embedding4], dim=0)
```

```python
all_embeddings = {}
with tqdm(total=len(nouns)) as pbar:
    for i in range(len(nouns)):
        embedding = extract_embedding(f'{nouns[i].name().split(".")[0]}: {vocab_def[i]}', model, tokenizer, device=device)
        all_embeddings[nouns[i].name()] = embedding.cpu()
        
        pbar.update(1)
```
