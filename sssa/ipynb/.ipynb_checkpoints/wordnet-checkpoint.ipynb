```python
filtered_nouns = list(filter(lambda x: x.max_depth()==1, nouns))
```

### WN class name counter

```python
import os 
import numpy as np 
from nltk.corpus import wordnet as wn
from collections import Counter

nouns = [x for x in wn.all_synsets(pos=wn.NOUN)]
nouns = list(map(lambda x: x.name().split('.')[0], nouns))
ct_nouns = Counter(nouns)
{k: v for k, v in ct_nouns.items() if v>2}
```
